```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
fig,axes = plt.subplots(2, 2, figsize = (10,8))

axes[0,0].set_title("Distribution of sepal length")
axes[0,0].hist(data['sepal_length'])

axes[0,1].set_title("Distribution of sepal width")
axes[0,1].hist(data['sepal_width'])

axes[1,0].set_title("Distribution of petal length")
axes[1,0].hist(data['petal_length'])

axes[1,1].set_title("Distribution of petal width")
axes[1,1].hist(data['petal_width'])

plt.show()
```

```python
fig,axes = plt.subplots(2, 2, figsize = (16,9))

axes[0,0].set_title("Distribution of sepal length")
sns.boxplot(y = "sepal_length", x = "species", data = data, ax=axes[0,0])

axes[0,1].set_title("Distribution of sepal width")
sns.boxplot(y = "sepal_width", x = "species", data = data, ax=axes[0,1])

axes[1,0].set_title("Distribution of petal length")
sns.boxplot(y = "petal_length", x = "species", data = data, ax=axes[1,0])

axes[1,1].set_title("Distribution of petal width")
sns.boxplot(y = "petal_width", x = "species", data = data, ax=axes[1,1])

plt.show()
```
